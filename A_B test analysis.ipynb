```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from google.colab import drive


from google.colab import drive
drive.mount("/content/drive")
%cd /content/drive/MyDrive/DA_study/


df = pd.read_csv("AB test query.csv")

df.head()
```

```python
#Analytic function for a/b test
def a_b_test_res(df, test_number) :
  #Choiced events
  conver_metrics = [
        "add_payment_info",
        "add_shipping_info",
        "begin_checkout",
        "new account",
        "session with orders"
    ]
  df_test = df[df['test'] == test_number]

  event_name_val = pd.pivot_table(
        df_test,
        index="event_name",
        columns="test_group",
        values="value",
        aggfunc="sum"
    )
  session_control  = event_name_val.loc["session", 1]
  session_test  = event_name_val.loc["session", 2]

  result_list  = []
#calculation of metrics
  for metric in conver_metrics:
        numeratic_control = event_name_val.loc[metric, 1]
        numeratic_test = event_name_val.loc[metric, 2]

        conversion_control = numeratic_control / session_control * 100
        conversion_test = numeratic_test / session_test *100

        metric_change = (conversion_test / conversion_control - 1) * 100

        z_test, p_value = sm.stats.proportions_ztest([numeratic_control, numeratic_test], [session_control, session_test])
#Creating df
        is_significant = p_value < 0.05
        result_list.append({
            "test_number" : test_number,
            "numerator_event" : metric,
            "metric" : f"{metric}/session",
            "denominator_event": "session",
            "numerator_control" : numeratic_control,
            "denominator_control": session_control,
            "conversion_control (%)" : conversion_control,
            "numerator_test" : numeratic_test,
            "denominator_test": session_test,
            "conversion_test (%)" : conversion_test,
            "metric change (%)" : metric_change,
            "z_stat" : z_test,
            "P-value" : p_value,
            "is_significant" : is_significant
        })
  analitic_test = pd.DataFrame(result_list)
  return analitic_test
#for all tests
full_df = []
for test_number in df['test'].unique():
      result_df = a_b_test_res(df, test_number)
      full_df.append(result_df)
      final_df = pd.concat(full_df, ignore_index=True)
      final_df = final_df.sort_values(by = "test_number").reset_index()
final_df.head(20)

```

```python
#Function to analysis in segments
def analysis_in_columns(df, test_number, segment_coll):
  all_results = []

  for segment in df[segment_coll].unique():
      segment_df = df[df[segment_coll] == segment]
      result_df = a_b_test_res(segment_df, test_number)
      result_df[segment_coll] = segment
      all_results.append(result_df)

  final_df = pd.concat(all_results, ignore_index=True)
  return final_df
```
